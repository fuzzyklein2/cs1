```python
def unique_values(s):
    v = list()
    for f in list_trip_files():
        df = pd.read_csv(f)
        for u in df[s].dropna().unique():
            v.append(u)
    v = list(set(v))
    # v.sort()
    if '' in v:
        v = v.remove('')
    return v

```
